```python
import pandas as pd
import numpy as np

import matplotlib
matplotlib.use('TkAgg')
import matplotlib.pyplot as plt

# cargar
df = pd.read_excel("con carrera.xlsx")

# columnas
t = df["Time"].values
y = df["Position\\Y Position In Current Wobj"].values
v = df["Speed In Current Wobj"].values

# === LIMPIEZA (clave) ===
mask_valid = np.isfinite(t) & np.isfinite(y) & np.isfinite(v)

t = t[mask_valid]
y = y[mask_valid]
v = v[mask_valid]

# === DEBUG RANGO ===
print("min Y:", y.min())
print("max Y:", y.max())

# === FILTRO intervalo en Y ===
mask = (y >= -100) & (y <= 100)

t_int = t[mask]
y_int = y[mask]
v_int = v[mask]

print("Cantidad en intervalo:", len(v_int))

# === PROMEDIO ===
if len(v_int) > 0:
    v_avg = np.mean(v_int)
    print(f"Velocidad promedio en Y ∈ [-100,100]: {v_avg:.4f}")
else:
    print("⚠️ No hay datos en ese intervalo")

# === GRAFICOS ===
plt.figure()

plt.subplot(2,1,1)
plt.plot(t, y)
plt.title("Posición Y vs Tiempo")
plt.grid()

plt.subplot(2,1,2)
plt.plot(t, v)
plt.title("Velocidad vs Tiempo")
plt.grid()

plt.tight_layout()
plt.show()
```
